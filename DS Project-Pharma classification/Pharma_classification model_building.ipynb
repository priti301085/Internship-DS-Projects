# **NULL VALUES TREATMENT**

# **TREATMENT:**(RETAIN|REPLACE|REMOVE)

## **Removing:**(For NULL Values 0% to 5%: Delete rows or columns with missing data if the percentage is low.

## **Imputing:**(REPLACING) For NULL Values 6% to 45% in Columns: Fill missing values with the mean, median, mode, or use a predictive model.

## **Removing :** For NULL Values 50% or More in Columns: Delete columns with high percentages of missing values, as they may not be reliable.

## **Flagging:** Mark missing values with a specific indicator for analysis

# **OUTLIER DETECTION**

```python
# ALL COLUMNS COATAIN OUTLIERS ...
import matplotlib.pyplot as plt
data.boxplot()
plt.show()
```

```python
import numpy as np

def datalu(data):
    for a in data.select_dtypes(include='number').columns:
        uq = data[a].quantile(0.75)
        lq = data[a].quantile(0.25)
        iqr = uq - lq

        upper_extreme = uq + (1.5 * iqr)
        lower_extreme = lq - (1.5 * iqr)

        print(f"Upper extreme of {a}: {upper_extreme}")
        print(f"Lower extreme of {a}: {lower_extreme}")

        # Outlier capping
        data[a] = np.where(data[a] < lower_extreme, lower_extreme,
                           np.where(data[a] > upper_extreme, upper_extreme, data[a]))
```

```python
datalu(data)
```

```python
plt.figure(figsize=(20,10))
data.boxplot()
plt.show()
# NOW THERE IS NO OUTLIERS
```

# **DIVIDING DATA INTO TRAINING NAD TESTING DATASET**

```python
# DIVIDING ORIGINAL DATA INTO X VARIABLES AND Y VARIABLE
x=data.iloc[:,1: ]
y=data.iloc[:,0]

# STANDARDISING DATA:(INDEPENDENT VARIABLE )
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
standard_data=ss.fit_transform(x)

# DIVIDING DATA INTO TRAINING AND TESTING SETS :
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(standard_data,y,test_size=0.3,random_state=1)
```

# **BALANCING INBALANCE CLASSES**

```python
from imblearn.over_sampling import SMOTE


smote = SMOTE(random_state=1)

X_resampled, y_resampled = smote.fit_resample(xtrain, ytrain)
```

```python
# Create a count plot to visualize the class distribution after SMOTE
plt.figure(figsize=(8, 6))
sns.countplot(x=y_resampled)
plt.title("Class Distribution After SMOTE")
plt.xlabel("Classes")
plt.ylabel("Count")
plt.show()
```

# **LOGISTIC REGRESSION MODEL**

```python
# IMPORT LOGISTIC REGRESSION:
from sklearn.linear_model import LogisticRegression

# CREATE INSTANCES:
lr=LogisticRegression()


# FIT TRAINING DATA TO TRAIN THE MODEL 
lr.fit(X_resampled, y_resampled)
```

```python
# TRAINING ACCURACY :
lr.score(X_resampled, y_resampled)
```

```python
# FOR MODEL TESTING :COMPUTING Y HAT FOR TEST DATA
y_pred = lr.predict(xtest)
y_pred
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
print("precision score =" ,precision_score(ytest,y_pred,average="weighted"))
print("recall score = ", recall_score(ytest,y_pred,average="weighted"))
print("f1 score = ",f1_score(ytest,y_pred,average="weighted"))
```
